```python
from imp_act import make

from imp_act.environments.recorder import Recorder

from imp_act.environments.rollout_plotter import RolloutPlotter
```

```python
env = make("ToyExample-v2")
```

```python
recorded_env = Recorder(env)

# record rollouts
for ep in range(10):

    obs = recorded_env.reset()
    done = False
    while not done:

        actions = [[0] * len(e["road_edge"].segments) for e in env.graph.es]

        obs, reward, done, info = recorded_env.step(actions)
```

```python
tape = recorded_env.tape

plotter = RolloutPlotter(env)

ep = 2 # episode to plot
plotter.plot(episode_data=tape[ep])
```
